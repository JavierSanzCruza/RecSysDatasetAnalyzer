```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Degree distributions

```python
file = "C:\\Users\\Javier\\Downloads\\ContentWiseImpressions\\data\\ContentWiseImpressions\\CW10M-CSV\\analysis\\pop-user-series.txt"
f = open(file)
x = []
y = []

header = True
for line in f:
    if header:
        header = False
    else:
        line.replace('\n','')
        data = line.split("\t")
        x.append(int(data[0]))
        y.append(int(data[1]))
        
```

```python
plt.plot(x,y,'o', color='blue', markersize=4)
plt.xlabel('Series', fontsize=18, family='Linux Libertine')
plt.xscale('log')
plt.xticks(fontsize=16, family='Linux Libertine')
plt.ylabel('Num. ratings', fontsize=18, family='Linux Libertine')
plt.yscale('log')
plt.yticks(fontsize=16, family='Linux Libertine')
plt.title('Series distribution', fontsize=22, family='Linux Libertine')

plt.show()
```

# Time series

```python
file = "C:\\Users\\Javier\\Downloads\\ContentWiseImpressions\\data\\ContentWiseImpressions\\CW10M-CSV\\analysis\\time-series.txt"
f = open(file)
times_series = []
series_series = []

header = True
for line in f:
    if header:
        header = False
    else:
        line.replace('\n','')
        data = line.split("\t")
        times_series.append(int(data[0]))
        series_series.append(int(data[1]))
```

```python
file = "C:\\Users\\Javier\\Downloads\\ContentWiseImpressions\\data\\ContentWiseImpressions\\CW10M-CSV\\analysis\\time-series-impressions.txt"
f = open(file)
times_impr = []
series_impr = []

header = True
for line in f:
    if header:
        header = False
    else:
        line.replace('\n','')
        data = line.split("\t")
        times_impr.append(int(data[0]))
        series_impr.append(int(data[1]))
```

```python
file = "C:\\Users\\Javier\\Downloads\\ContentWiseImpressions\\data\\ContentWiseImpressions\\CW10M-CSV\\analysis\\time-users.txt"
f = open(file)
times_users = []
series_users = []

header = True
for line in f:
    if header:
        header = False
    else:
        line.replace('\n','')
        data = line.split("\t")
        times_users.append(int(data[0]))
        series_users.append(int(data[1]))
```

```python
fig, (axseries,aximpr,axuser) = plt.subplots(3,1, figsize = (20,20))
plt.xticks(fontsize=16, family='Linux Libertine')
plt.yticks(fontsize=16, family='Linux Libertine')

axseries.plot(series_series,times_series,'o', color='black', markersize=1)
axseries.set_xlabel('Timestamp', fontsize=18, family='Linux Libertine')
axseries.set_ylabel('Series', fontsize=18, family='Linux Libertine')
axseries.set_title('Time distribution (series)', fontsize=22, family='Linux Libertine')

aximpr.plot(series_impr,times_impr, 'o', color='black', markersize=1)
aximpr.set_xlabel('Timestamp', fontsize=18, family='Linux Libertine')
aximpr.set_ylabel('Series', fontsize=18, family='Linux Libertine')
aximpr.set_title('Time distribution (series w. impressions)', fontsize=22, family='Linux Libertine')

axuser.plot(series_users,times_users, 'o', color='black', markersize=1)
axuser.set_xlabel('Timestamp', fontsize=18, family='Linux Libertine')
axuser.set_ylabel('Series', fontsize=18, family='Linux Libertine')
axuser.set_title('Time distribution (users)', fontsize=22, family='Linux Libertine')

plt.show()
```
